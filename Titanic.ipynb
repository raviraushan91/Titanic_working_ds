```python
data.drop('Cabin',axis=1,inplace=True)
```

```python
data['Age'].fillna(data['Age'].median(),inplace=True)
```

```python
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data1=pd.get_dummies(data,columns=['Sex','Embarked'],drop_first=True)
```

```python
data1.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
data1
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
 num_colms=['Age','Fare']
```

```python
ssc=StandardScaler()
```

```python
data1[num_colms]=ssc.fit_transform(data1[num_colms])
```

```python
data1.head()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(data1.corr(),annot=True)
```

```python
for i, col in enumerate(num_colms):
    plt.subplot(1, 2, i+1)
    sns.boxplot(y=data1[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
for col in num_colms:
    Q1 = data1[col].quantile(0.25)
    Q3 = data1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    data1 = data1[(data1[col] >= lower) & (data1[col] <= upper)]
```
